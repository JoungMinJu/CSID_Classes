```python
#2차원, 100개의 랜덤 샘플 생성=샘플 X
#2차원의 고정된 점 하나를 생성 = 기준점 z
#기준점으로부터 샘플까지의 거리를 계산
#계산한 거리를 기준으로, 일정 거리 이상이 되는 애들을 따로 표시

X=np.random.rand(100,2) 
#100개의 샘플들이 행으로 들어가있다! 100개의 행벡터

#아래처럼 표현해도됨

N=100 #샘플의 개수
D=2
X=np.random.rand(N,D)
z=np.array([0.3,0.4])


#거리 계산하는 두 가지 방법 1.for문 사용 2.for문 사용안함
##두가지 시간 비교도 할 것이다요

t=TicToc() #타이머

#1.for문 사용
t.tic()
dist1=np.zeros(N)  #distance라는 벡터 생성
for n in range(0,N):
    #X의 n번째 샘플을 z와 거리 비교
    dist1[n]=np.sqrt((X[n,0]-z[0])**2+(X[n,1]-z[1])**2)
t.toc()


#그림그리기
plt.figure()
plt.scatter(X[:,0],X[:,1], s=10,facecolor='gray') #샘플들! s=마커의 크기 설정
plt.scatter(z[0],z[1],marker='x', facecolor='red') #기준점
    
d_star=0.5 #거리의 기준
id_long=np.where(dist1>=d_star)

plt.figure()
plt.scatter(X[id_long,0],X[id_long,1], s=50,facecolor='cyan') #기준거리보다 멀어진 애들 하이라이트
plt.scatter(X[:,0],X[:,1], s=10,facecolor='gray') 
plt.scatter(z[0],z[1],marker='x', facecolor='red') 



#거리구하기 2. for문 사용 안함
t.tic()
Z=np.matlib.repmat(z.reshape(1,2),N,1) #X랑 동일한 shape으로 생성!
#z는 기준점이 담겨있음. 
dist2=np.sqrt(np.sum((X-Z)**2, axis=1)) #for문 없이 한 줄로 거리를 구함
t.toc()


#1번이랑 2번이랑 같은지
print(dist1-dist2)
```
