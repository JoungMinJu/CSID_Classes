```python
import numpy as np
import matplotlib.pyplot as plt

#1차원 입력(+bias 포함해서 2차원)
x=2*np.random.rand(100,1)

#1차원 출력= theta_0+theta_1*x
y=4+5*x+np.random.rand(100,1) #bias=4, theta_1=5, 뒤에 noise를 줘서 분포나타나게

plt.figure(figsize=(5,3))
plt.scatter(x,y,s=10)
plt.xlabel('x')
plt.ylabel('y')
```

```python
# analytic, normal equation

# design matrix = 샘플 수 x 입력 수 (차원의 수)
X = np.hstack([np.ones((100,1)), x]) # 100 x 2
#1로 가득찬 열벡터+X열벡터
# X = x <--bia 텀을 포함하지 않고 regression하는 경우

# \hat{theta} = (X^T X)^{-1} X^T y
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
#dot쓰는건 행렬곱

x_new = np.linspace(0,2,10).reshape(-1,1) #컬럼벡터로 만들었음
# y_new = theta[0,0]*x_new <--bias텀을 포함하고 싶지 않은경우
y_new = theta[0,0] + theta[1,0]*x_new

plt.figure(figsize = (5,3))
plt.scatter(x,y,s=10)
plt.plot(x_new,y_new)
plt.xlabel('x')
plt.ylabel('y')
```
