```python
#scatter사용한 버전


import numpy as np 
import numpy.matlib 
import matplotlib.pyplot as plt

# np.random.rand = uniform random generation, [0, 1]
# np.random.randn = normal random generation, N(0, 1)
r1 = 10*(np.random.rand(2,2) - 0.5) # [-1, 1]  #로테이션의 의미
#실행할 때마다 random으로 ! 정규분포를 찌그러뜨린다.
r2 = 10*(np.random.rand(2,2) - 0.5) # [-1, 1]
r3 = 10*(np.random.rand(2,2) - 0.5) # [-1, 1]

t1 = 50*(np.random.rand(1,2) - 0.5)
t2 = 50*(np.random.rand(1,2) - 0.5)
t3 = 50*(np.random.rand(1,2) - 0.5)

X1 = np.random.randn(1000,2)
X1 = X1.dot(r1) + t1 #dot은 행렬 곱
#(1000*2)*(2*2) 행렬곱 한 것 
#t1을 더하는 것은 중심 위치도 옮기려고! 

X2 = np.random.randn(500,2)
X2 = X2.dot(r2) + t2

X3 = np.random.randn(500,2)
X3 = X3.dot(r3) + t3

plt.figure(figsize = (5,5))
plt.scatter(X1[:,0],X1[:,1],s=10)
#(x축, y축, s=마커의 크기 설정, c=마커의 색상 설정)
plt.scatter(X2[:,0],X2[:,1],s=10)
plt.scatter(X3[:,0],X3[:,1],s=10)


# unlabel data
X = np.vstack([X1, X2, X3])
np.random.shuffle(X)

plt.figure(figsize = (5,5))
plt.scatter(X[:,0],X[:,1],s=10, color='gray')
#클러스터링 알고리즘에는 이렇게 레이블 없이 들어간다
```

```python
#parameter
K=3  #클러스터의 개수
maxiter=10 #최대 반복할 횟수 지정

#nomalization
Xmin=np.min(X,axis=0)
Xmax=np.max(X, axis=0)
X=(X-Xmin)/(Xmax-Xmin)

plt.figure(figsize = (5,5))
plt.scatter(X[:,0],X[:,1],s=10, color='gray') #데이터 범위가 0~1로 고정됐음을 확인할 수 있음

#초기화
(N,D)=X.shape
#N : 샘플의 수 
#D : 입력의 차원
#K : cluster의 개수
mu_traj=np.zeros((K,D,maxiter)) #iteration이 돌아가면서 뮤가 바뀌는데 ,그거 ㄹ저장하기 위해 3차원 array만듬
dist=np.zeros((N,K))

#k-means 본론

#0. mu를 랜덤으로 초기화
mu=np.random.rand(K,D) #unifor random,[0,1]사이에서 균일한 분포로 랜덤을 생성 (scailing할 필요가 없음)
# mu1=mu[0,:] 1군집에 대한 mu (평균)
# mu2=mu[1,:]
# mu3=mu[2,:] 
# ..
mu_traj[:,:,0]=mu #처음 뮤 넣고


for iter in range(maxiter):
    #1. 전체 데이터에 대해서 각각의 뮤로부터 거리를 게산
    for k in range(K):
        #dist의 차원은 [샘플의수, 클러스터의 수]
        dist[:,k]=np.sqrt(np.sum((X-np.matlib.repmat(mu[k,:],N,1))**2,axis=1))
#       B = repmat(A,n)은 A의 복사본 n개를 포함하는 배열을 행과 열 차원으로 반환합니다. A가 행렬인 경우 B의 크기는 size(A)*n입니다.
        #교수님 코드에서는 np.sqrt가 없었음.
       
        
    #dist의 1,1은 X[0,:]과 mu[0,:] 사이의 거리 (첫번쨰 샘플과 첫번째 뮤까지 거리값)
    #dist 1,2는 첫번째 샘플과 두 번째 뮤까지의 거리 값...
    

    #2. 계산한 거리로부터 어떤 뮤에 제일 가까운지 계산 (r을 계산)
    label=np.argmin(dist, axis=1) #최소값이 나타나는 인덱스 반환
    print(dist.shape)
    print(label.shape)



    #3. 각각의 뮤에 제일 가깝다고 계산된 샘플들끼리 평균을 게산 => 뮤에 업데이트 (뮤를 계산)
    for k in range(K):
        mu[k,:]=np.mean(X[label==k,:], axis=0)
    #X[label==k,:] 레이블이 k인 샘플들!
    mu_traj[:,:,iter]=mu

    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.scatter(X[:,0],X[:,1],s=10,color='gray')#레이블이 없는 그림

    plt.subplot(1,2,2)
    for k in range(K):
        plt.scatter(X[label==k,0],X[label==k,1],s=10) #레이블이 있는 그림
        
    plt.gca().set_prop_cycle(None) #색상 로테이션을 초기화
    #plt.gca()현재 Axes(축) 객체를 반환
    #matplotlib으 ㄴ자동으로 선에 색상을 할당하는데 각 라인의 색상을 제어하기 위해
    #matplotlib.axes.Axes.set_prop_cycle()메서드를 사용한다.
    
    for k in range(K):
        plt.scatter(mu[k,0],mu[k,1],marker='x',s=50)
# plt.scatter(mu_traj[k,0,0:2], mu_traj[k,1,0:2],marker='x')
```

```python
#parameter
K=3  #클러스터의 개수
maxiter=10 #최대 반복할 횟수 지정

#nomalization
Xmin=np.min(X,axis=0)
Xmax=np.max(X, axis=0)
X=(X-Xmin)/(Xmax-Xmin)

plt.figure(figsize = (5,5))
plt.scatter(X[:,0],X[:,1],s=10, color='gray') #데이터 범위가 0~1로 고정됐음을 확인할 수 있음

#초기화
(N,D)=X.shape
#N : 샘플의 수 
#D : 입력의 차원
#K : cluster의 개수
mu_traj=np.zeros((K,D,maxiter)) #iteration이 돌아가면서 뮤가 바뀌는데 ,그거 ㄹ저장하기 위해 3차원 array만듬
dist=np.zeros((N,K))

#k-mean 본론

#0. mu를 랜덤으로 초기화
mu=np.random.rand(K,D) #unifor random,[0,1]사이에서 균일한 분포로 랜덤을 생성 (scailing할 필요가 없음)
# mu1=mu[0,:]
# mu2=mu[1,:]
# mu3=mu[2,:]
# ..
mu_traj[:,:,0]=mu #처음 뮤 넣고


for iter in range(maxiter):
    #1. 전체 데이터에 대해서 각각의 뮤로부터 거리를 게산
    for k in range(K):
        dist[:,k]=sqrt(np.sum((X-np.matlib.repmat(mu[k,:],N,1))**2,axis=1))
     
    #dist의 1,1은 X[0,:]과 mu[0,:] 사이의 거리 (첫번쨰 샘플과 첫번째 뮤까지 거리값)
    #dist 1,2는 첫번째 샘플과 두 번째 뮤까지의 거리 값...
    

    #2. 계산한 거리로부터 어떤 뮤에 제일 가까운지 계산 (r을 계산)
    label=np.argmin(dist, axis=1) #최소값이 나타나는 인덱스 반환
    #애를 pne-hot 벡터로 표현하는 것도 할 수 있음. (이거 미리 올려주신 코드에 이 부분 있대)


    #3. 각각의 뮤에 제일 가깝다고 계산된 샘플들끼리 평균을 게산 => 뮤에 업데이트 (뮤를 계산)
    for k in range(K):
        mu[k,:]=np.mean(X[label==k,:], axis=0)
    #X[label==k,:] 레이블이 k인 샘플들!
    mu_traj[:,:,iter]=mu

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],s=10,color='gray')

plt.subplot(1,2,2)
for k in range(K):
    plt.scatter(X[label==k,0],X[label==k,1],s=10)
plt.gca().set_prop_cycle(None) #색상 로테이션을 초기화
for k in range(K):
 #       plt.scatter(mu[k,0],mu[k,1],marker='x',s=50)
    plt.plot(mu_traj[k,0,:], mu_traj[k,1,:],marker='x')


#최종 결과만 출력!
```
