```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

N=100 #전체 샘플 백개
samples = np.random.rand(N)
#N(0,1)을 따르는 랜덤 샘플 백개(N)를 생성
#N(theta, 1)

plt.figure()
plt.hist(samples)
plt.xlabel('sample range')
plt.ylabel('sample count')
```

```python
K=100
theta = np.linspace(-1,1, K)
#세타의 후보군! [-1, +1] 사이의 실수들! 100개로 쪼개서! 즉, 총 백개의 후보군

#likelihood 분포에 넣어보기
prob_likelihood = np.zeros(K)
for k in range(K):
    prob_likelihood[k] = np.prod(multivariate_normal.pdf(samples,mean=theta[k], cov= 1))
#     print('원래는', multivariate_normal.pdf(samples, mean=theta[k], cov=1))
#     print("이거다", prob_likelihood[k])
    
# print(theta)
plt.figure()
plt.plot(theta, prob_likelihood)


#likelihood가 의미하는 것? 
#각 세타에 대해서 전체 샘플의 probability 구해서 점 찍어서 그거 연결한것
#아마 그 최대의 지점은 샘플들의 산술평균일 것

print()
theta_mle = theta[np.argmax(prob_likelihood)]
print('최대만드는 세타', theta_mle)
print('샘플들의 평균', np.mean(samples))
#비슷한 값이 나온다!
```
